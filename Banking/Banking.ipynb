```python
#List to store the account_holder_name
account_holder_name=[]
#Dictionary to store the details
'''Key is name
   Value is Account Balance, Contact Number'''
account_details={}
#Password to access admin
admin_password="admin123!@"
def admin(password):
    #check for password
    if password!=admin_password:
        s="You have entered wrong password"
        return s
    #List of Choices
    choice=int(input("\nEnter the choice:\n1. Name of all the Account_holder\n2. Account Balance\n3. Modify Balance\n4. Modify contact number\n5. Enter the new Details\n6. Exit\n"))
    #Iterate till admin want to exit
    while(choice!=6):
        #Condition for choice==1
        if choice==1:
            #If no account_details have been added
            if len(account_holder_name)==0:
                print("\nNo account details have been added. Requested to add first\n")
            else:
                print("\nThe name of all the account holder\n")
                for i in account_holder_name:
                    print(i)
        
        #Condition for choice==2
        elif choice==2:
            #If no account_details have been added
            if len(account_holder_name)==0:
                print("\nNo account details have been added. Requested to add first\n")
            else:
                name=input("\nEnter the name of Account holder to get the balance\n")
                if name not in account_details:
                    print("\nThe name of Account Holder not exist.\n")
                else:
                    #0th index have the account balance
                    print("\n The account Balance is: ",account_details[name][0])
        
        #Condition for choice==3
        elif choice==3:
            name=input("\nEnter the name of Account holder to modify the balance\n")
            #if name(key) not in our dictionary
            if name not in account_details:
                print("\nThe name of Account Holder not exist.\n")
            else:
                value=input("\nEnter the new balance\n")
                account_details[name][0]=value
                option=int(input("\nThe balance is modified, you can check the new balance\n1. Check balance\n2. Previous Menu\n"))
                if option==1:
                        #0th index have the account balance
                        print("\nThe new balance is",account_details[name][0])
                       
        #Condition for choice==4
        elif choice==4:
            name=input("\nEnter the name of Account holder to modify the Contact Number\n")
            #if name(key) not in our dictionary
            if name not in account_details:
                print("\nThe name of Account Holder not exist.\n")
            else:
                value=input("\nEnter the new Number\n")
                account_details[name][1]=value
                option=int(input("\nThe balance is modified, you can check the new Contact Number\n1. Check balance\n2. Previous Menun\n"))
                if option==1:
                    #1st index have the Contact details
                    print("\nThe new Contact Number\n",account_details[name][1])
        
        #Condition for choice==5
        elif choice==5:
            name=input("\nEnter the name of Account holder\n")
            #If name(key) already present in dictionary
            if name in account_details:
                print("\nAccount holder name already exist. Retry\n")

            else:
                balance=int(input("\nEnter the opening account balance\n"))
                phone=int(input("\nEnter the contact detail\n"))
                detail=[]
                detail.append(balance)
                detail.append(phone)
                account_details[name]=detail
                account_holder_name.append(name)
                print("\nAccount details added.\n")
        
        #Condition for choice==6
        elif choice==6:
            break
        
        
        else:
            print("\nWrong choice selection. Retry\n")
        choice=int(input("\nEnter the choice:\n1. Name of all the Account_holder\n2. Account Balance\n3. Modify Balance\n4. Modify contact number\n5. Enter the new Details\n6. Exit\n"))
```

```python
def user(name):
    #if name not in our record
    if name not in account_details:
        print("\nThe name of Account Holder not exist.\n")
        return
    #List of choice for the user
    choice=int(input("\nEnter the choice:\n1. Account Balance\n2. Contact Number\n3. Exit\n"))
    #Iterate till user want to exit
    while choice!=3:
        if choice==1:
                print("\nThe Account Balance is: ",account_details[name][0],"\n")
        elif choice==2:
                print("\nThe Contact Number is: ",account_details[name][1],"\n")
        elif choice==3:
            break
        else:
            print("\nWrong choice selection. Retry\n")
        choice=int(input("\nEnter the choice:\n1. Account Balance\n2. Contact Number\n3. Exit\n"))
```

```python
def bank():
    print('\t "Welcome to Our Bank"\n')
    choice=int(input("\nEnter your choice\n1. Admin\n2. User\n3. Exit\n"))
    while choice!=3:
        if choice==1:
            password=input("\nEnter the password\n")
            admin(password)
        elif choice==2:
            name=input("\nEnter the name\n")
            user(name)
        elif choice==3:
            return
        else:
            print("\nWrong choice selection. Retry\n")
        choice=int(input("\nEnter your choice\n1. Admin\n2. User\n3. Exit\n"))
```
